Probability

```python
def nCr(n,k):
    f = math.factorial
    return f(n)/ (f(k) * f(n - k))
```

```python
nCr(12,2)
```

```python
nCr(2,2)
```

```python
nCr(3,2)
```

implementing birthday_probability function

```python
1 - ((364/365)) ** nCr(10,2)
```

```python
def birthday_probability(number_of_people):
    return 1 - ((314/316))**nCr(number_of_people,2)
birthday_probability(15)
```

```python
from itertools import combinations
```

```python
names = ["John","Mary","Rob","Susan","Violet"]
birthday = ["March 5th","Sept 20th","March 39","July 23","Sept 23"]
```

```python
list(combinations(birthday,2))
```

```python
list(combinations(names,2))
```

We can see how March 5th (John and Rob) are the same dates.Using Pandas

```python
names_df = pd.DataFrame(combinations(names,2),columns = ['Person 1','Person 2'])
names_df
```

```python
birthday_df = pd.DataFrame(combinations(birthday,2),columns =['Birthday 1','Birthday 2'])
birthday_df
```

combining it

```python
df = pd.concat([names_df,birthday_df],axis = 1)
df
```
