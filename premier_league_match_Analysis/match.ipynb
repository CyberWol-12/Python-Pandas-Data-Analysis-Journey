analysis

Whats the average of FTAG and HTAG

```python
(df['FTAG'] + df['HTAG']).mean()
```

Create a new column Total_goals

```python
df['Total_goals'] = (df['FTAG'] + df['HTAG']).mean()
```

```python
df.head()
```

Calculate average goal per seasons

```python
goal_per_seasons = df.groupby('Season')['Total_goals'].mean().sort_index()
goal_per_seasons
```

Whats the team with most E0 wins?

```python
df.loc[df['Div'] == 'E0'].groupby('AwayTeam')['Season'].size().sort_values(ascending = False).head()
```

```python
df.groupby('AwayTeam').apply(lambda rows: (rows['Div'] == 'E0').sum()).sort_values(ascending = False).head()
```

Whats the team with most goals scored at home?

```python
df.head()
```

whats the team the received the least amount of goals while playing ast home

```python
homes_games_per_team = df.groupby('HomeTeam')[['HomeTeam','HTAG']].agg({
    'HomeTeam': 'size',
    'HTAG': 'sum',
}).rename(
    columns = {'HomeTeam':'Total_goals','HTAG':'goal_received'}
).sort_values(by = ['HomeTeam','goal_received'],ascending = [False,True])
homes_games_per_team
```
