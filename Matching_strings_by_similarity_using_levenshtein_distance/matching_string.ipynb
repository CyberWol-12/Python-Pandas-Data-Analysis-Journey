```python
A = ['Apple','Alphabet','Microsoft']
B = ['MSFT','Alphabet/Google','Apple inc.']
```

```python
companies = list(itertools.product(A,B))
companies
```

```python
for c1,c2 in companies:
    ratio = fuzz.partial_ratio(c1,c2)
    print(f"{c1} > {c2}: {ratio}")
```
